```r
for (c1 in compvalsA) {
    for (c2 in compvalsB) {
        if (c1==c2){
                    next
                }
        finnameFG<-paste('HQ_G2_beds/61819_','C',toString(c1),'C',toString(c2),'G2_HQ_gr8_FG.bed', sep='')
        print(finnameFG)
        mybedFG = read.csv(finnameFG, sep='\t', header=FALSE, col.names=c('chr', 'start', 'end'))

        flush.console()
        job <- submitGreatJob(mybedFG, bg=mybedBG, request_interval=10)

        tb <- getEnrichmentTables(job, ontology = c('GO Molecular Function', 'GO Biological Process','Mouse Phenotype', 'Human Phenotype','MSigDB Oncogenic Signatures','Placenta Disorders', 'Disease Ontology', 'MGI Expression: Detected', 'MSigDB Perturbation' ))

        tbnames <- names(tb)
         for (i in seq(length(names(tb)))){
            nameoftable<-tbnames[i]
            print(nameoftable)
            foutname<-paste('61819results/C',c1,'/61819_G2_HQ_gr8_','C',toString(c1),'C',toString(c2),nameoftable,'.csv', sep='')
            cut <- tb[[i]]$Hyper_Adjp_BH < 0.01
            write.csv(tb[[i]][cut,], file =foutname)
        }



        do_the_GO <- tb[[2]]$Hyper_Adjp_BH < 1e-2
        gos_to_do <- tb[[2]]$ID[do_the_GO]

    }




}
```
