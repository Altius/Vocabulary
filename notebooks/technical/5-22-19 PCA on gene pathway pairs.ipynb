```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
GO_finname = '50719results/C8C10/50719_G2_G1bg_gr8_C8C10GO Biological Process.csv'
```

```python
GO_DF = pd.read_csv(GO_finname)
```

```python
CompA = 'C8'
CompB = 'C10'
```

```python
gene_pathway_table = []
for i in range(GO_DF.shape[0]):
    GOID = GO_DF.loc[i].ID
    finname = '50719results/'+CompA+CompB+'/50619_G2_gr8_'+CompA+CompB+'_'+GOID+'bioproc.csv'
    regionsDF = pd.read_csv(finname)
    genelist = list(set(regionsDF.gene.values))
    gene_pathway_table.append(genelist)
```

```python
gene_pathway_table = np.array(gene_pathway_table)
```

```python
all_gene_set = list(set(np.concatenate(gene_pathway_table).tolist()))
```

```python
path_gene_matrix = []
for i in range(len(gene_pathway_table)):
    path_gene_matrix.append(np.in1d(all_gene_set, gene_pathway_table[i]))
```

```python
path_gene_matrixDF = pd.DataFrame(path_gene_matrix, columns =all_gene_set ) 
```

```python
path_gene_matrixDF['pathways'] = GO_DF['name']
```

```python
path_gene_matrixDF.set_index('pathways', inplace=True)
```

```python
path_gene_matrixDF.shape
```

```python
path_gene_matrixDF.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit(path_gene_matrixDF)
```

```python
print(pca.explained_variance_ratio_)  
```

```python
np.sum(pca.explained_variance_ratio_[0:8])
```

```python
np.sum(pca.explained_variance_ratio_[0:12])
```

```python
np.sum(pca.explained_variance_ratio_[0:16])
```

```python
PC_for_things = pca.transform(path_gene_matrixDF)
```

```python
PC_for_things
```

```python
PC_for_things.shape
```

```python
PC_for_thingsDF = pd.DataFrame(PC_for_things, index=path_gene_matrixDF.index, columns=['PC' +str(i+1) for i in range(PC_for_things.shape[0])])
```

```python
PC_for_thingsDF
```

```python
plt.clf()
plt.figure(figsize=(20,20))
plt.plot(PC_for_thingsDF.PC1.values, PC_for_thingsDF.PC2.values, 'ok')
for i in range(len(PC_for_thingsDF.PC1.values)):
    color_to_use = 'k'
    if (('nephr' in PC_for_thingsDF.index.values[i]) or ('renal' in PC_for_thingsDF.index.values[i]) or ('kidney' in PC_for_thingsDF.index.values[i])):
        color_to_use = 'r'
    elif (('neuro' in PC_for_thingsDF.index.values[i]) or ('brain' in PC_for_thingsDF.index.values[i]) or ('spin' in PC_for_thingsDF.index.values[i]) or ('eye' in PC_for_thingsDF.index.values[i]) or ('lens' in PC_for_thingsDF.index.values[i]) or ('nerv' in PC_for_thingsDF.index.values[i])):
        color_to_use ='b'
    plt.text(PC_for_thingsDF.PC1.values[i], PC_for_thingsDF.PC2.values[i], PC_for_thingsDF.index.values[i], color=color_to_use)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('52219_PC1PC2.pdf')
plt.close()
```
