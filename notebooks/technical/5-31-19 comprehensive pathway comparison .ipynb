```python
finname = 'gene_stats_all_WM20181115.txt'
geneDF = pd.read_csv(finname, sep='\t')
```

```python
geneDF.head()
```

```python
gene_pathway_tableCACB = []
for i in range(GO_DF_CACB.shape[0]):
    if not cutCACB[i]:
        continue
    GOID = GO_DF_CACB.loc[i].ID
    finname = '51419_results/'+compA+compB+'/50619_G2_gr8_'+compA+compB+'_'+GOID+'bioproc.csv'
    regionsDF = pd.read_csv(finname)
    genelist = list(set(regionsDF.gene.values))
    gene_pathway_tableCACB.append(genelist)
```

```python
all_gene_setCACB = list(set(np.concatenate(gene_pathway_tableCACB).tolist()))
```

```python
geneDF.loc[all_gene_setCACB].top_comp.value_counts().sort_index().plot('bar', color='k')
plt.xlabel('component')
plt.ylabel('N genes per wouter top comp association')
plt.title('genes from G2 '+compA+compB+' enriched GREAT pathways')
plt.savefig('53119_'+compA+compB+'_gene_topcomps.pdf', bbox_inches='tight')
plt.close()

( geneDF.top_comp.value_counts().sum() * geneDF.loc[all_gene_setCACB].top_comp.value_counts().sort_index() / (geneDF.top_comp.value_counts() * 16)).plot('bar', color='k')    
plt.xlabel('component')
plt.ylabel('N genes per wouter top comp association (normed)')
plt.title('genes from G2 '+compA+compB+' enriched GREAT pathways')
plt.savefig('53119_'+compA+compB+'_gene_topcomps_normed.pdf', bbox_inches='tight')
plt.close()
```

```python
gene_pathway_tableCBCA = []
for i in range(GO_DF_CBCA.shape[0]):
    if not cutCBCA[i]:
        continue
    GOID = GO_DF_CBCA.loc[i].ID
    finname = '51419_results/'+compB+compA+'/50619_G2_gr8_'+compB+compA+'_'+GOID+'bioproc.csv'
    regionsDF = pd.read_csv(finname)
    genelist = list(set(regionsDF.gene.values))
    gene_pathway_tableCBCA.append(genelist)

if len(gene_pathway_tableCBCA) > 0:
    all_gene_setCBCA = list(set(np.concatenate(gene_pathway_tableCBCA).tolist()))


    geneDF.loc[all_gene_setCBCA].top_comp.value_counts().sort_index().plot('bar', color='k')
    plt.xlabel('component')
    plt.ylabel('N genes per wouter top comp association')
    plt.title('genes from G2 '+compB+compA+' enriched GREAT pathways')
    plt.savefig('53119_'+compB+compA+'_gene_topcomps.pdf', bbox_inches='tight')
    plt.close()
    plt.clf()
    ( geneDF.top_comp.value_counts().sum() * geneDF.loc[all_gene_setCBCA].top_comp.value_counts().sort_index() / (geneDF.top_comp.value_counts() * 16)).plot('bar', color='k')    
    plt.xlabel('component')
    plt.ylabel('N genes per wouter top comp association (normed)')
    plt.title('genes from G2 '+compB+compA+' enriched GREAT pathways')
    plt.savefig('53119_'+compB+compA+'_gene_topcomps_normed.pdf', bbox_inches='tight')
    plt.close()
    
else:
    print('nothing to show')
```
