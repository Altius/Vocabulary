```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
combos = ['C1-C2', 'C3-C5', 'C3-C8', 'C3-C9', 'C3-C10', 'C3-C12', 'C3-C15', 'C5-C7', 'C8-C9', 'C8-C10', 'C5-C15']
```

```python
combos[0].split('-')
```

```python
for combo in combos:
    print(combo)
    Ci = combo.split('-')[0]
    Cj = combo.split('-')[1]
    G2ij_finname = '53119_exclusive_'+Ci+Cj+'G2_HQ_top25.csv'
    G2ji_finname = '53119_exclusive_'+Cj+Ci+'G2_HQ_top25.csv'
    G1i_finname = '53119_exclusive_'+Ci+'G1solo_HQ_top25.csv'
    G1j_finname = '53119_exclusive_'+Cj+'G1solo_HQ_top25.csv'
    G2ij_joint_finname = '53119_joint_'+Ci+Cj+'G2_HQ_top25.csv'

    lastone = 0
    plt.clf()
    plt.figure(figsize=(8,25))
    
    #print(Ci + ' solo G1')

    #print(pd.read_csv(G1i_finname, sep='\t')['name'].values)
    paths = pd.read_csv(G1i_finname, sep='\t')['name'].values

    plt.text(0, lastone, Ci+' solo', size=12, color='r')
    lastone-=2
    plt.plot(0, lastone,[0]*len(paths), -1*np.arange(len(paths)), 'ow')
    for i, path in enumerate(paths):
        plt.text(0, lastone-i, path, size=8)

    lastone += -1*len(paths)
    

    #print(Ci + ' primary '+Cj+ ' secondary '+' G2 exclusive')
    #print(pd.read_csv(G2ij_finname, sep='\t')['name'].values)

    paths = pd.read_csv(G2ij_finname, sep='\t')['name'].values

    lastone-=2
    plt.text(0, lastone,Ci + ' primary '+Cj+ ' secondary '+' G2 exclusive', size=12, color='r')
    lastone-=2
    plt.plot(0, lastone, [0]*len(paths), lastone+-1*np.arange(len(paths)), 'ow')
    for i, path in enumerate(paths):
        plt.text(0, lastone-i, path, size=8)
    lastone += -1*len(paths)


    #print(pd.read_csv(G2ij_joint_finname, sep='\t')['name'].values)

    paths = pd.read_csv(G2ij_joint_finname, sep='\t')['name'].values

    lastone-=2
    plt.text(0, lastone,Ci + ' '+Cj+ ' '+'G2 joint', size=12, color='r')
    lastone-=2
    #print(Ci + ' '+Cj+ ' '+'G2 joint')

    plt.plot([0]*len(paths), lastone+-1*np.arange(len(paths)), 'ow')
    for i, path in enumerate(paths):
        plt.text(0, lastone-i, path, size=8)
    lastone += -1*len(paths)

    paths = pd.read_csv(G2ji_finname, sep='\t')['name'].values
    #print(Cj + ' primary '+Ci+ ' secondary '+' G2 exclusive')
    lastone-=2
    plt.text(0, lastone,Cj + ' primary '+Ci+ ' secondary '+' G2 exclusive', size=12, color='r')
    lastone-=2
    #print(pd.read_csv(G2ji_finname, sep='\t')['name'].values)
    
    plt.plot([0]*len(paths), lastone+-1*np.arange(len(paths)), 'ow')
    for i, path in enumerate(paths):
        plt.text(0, lastone-i, path, size=8)
    lastone += -1*len(paths)
    
    paths = pd.read_csv(G1j_finname, sep='\t')['name'].values

    #print(Cj + ' solo G1')
    lastone-=2
    plt.text(0, lastone, Cj+' solo', size=12, color='r')
    lastone-=2

    #print(pd.read_csv(G1j_finname, sep='\t')['name'].values)    

    plt.plot([0]*len(paths), lastone+-1*np.arange(len(paths)),'ow')
    for i, path in enumerate(paths):
        plt.text(0, lastone-i, path, size=8)
    lastone += -1*len(paths)
    
    plt.axis('off')
    plt.savefig('60619'+Ci+Cj+'pathway_venn.pdf', bbox_inches='tight')
    
    plt.show()
```
