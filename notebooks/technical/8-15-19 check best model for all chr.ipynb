```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DF = pd.read_csv('81519_coarse_screen_sweep_results_total_allchr.txt', sep=' ', header=None, names=['f1','f2','f3','f4','f5','costhresh','uniq','cov'])
```

```python
DF['uniq_rat'] = DF['uniq']/DF['uniq'].max()
```

```python
DF['cov_rat'] = DF['cov']/DF['cov'].max()
```

```python
DF['total_score'] = DF['uniq_rat'] * DF['cov_rat']
```

```python
DF['unnormed_score'] = DF['uniq'] * DF['cov']
```

```python
DF.sort_values(by='total_score',ascending=False).head(12)
```

```python
DF.sort_values(by='total_score',ascending=False).tail(12)
```

```python
import matplotlib.pyplot as plt
```

```python

DF.sort_values(by='total_score',ascending=False)['total_score'].plot.bar(color='k')
plt.ylim([0.4, 0.82])
plt.ylabel('total score (f_coverage * f_unique)')
plt.xlabel('model number')
```

```python
plt.clf()
plt.plot(DF['cov'], DF['uniq'], 'ok')
plt.xlabel('fraction DHSs covered')
plt.ylabel('fraction unique DHSs')
plt.show()
```

```python
uniq_x = DF.sort_values(by='total_score',ascending=False).head(3).uniq.values
```

```python
cov_x = DF.sort_values(by='total_score',ascending=False).head(3)['cov'].values
```

```python
plt.clf()
plt.plot(DF['cov'], DF['uniq'], 'ok')
plt.plot([cov_x[0]], [uniq_x[0]], '*r', ms=10, fillstyle='none')
plt.plot([cov_x[1]], [uniq_x[1]], '*b', ms=10, fillstyle='none')
plt.plot([cov_x[2]], [uniq_x[2]], '*g', ms=10, fillstyle='none')
plt.xlabel('fraction DHSs covered')
plt.ylabel('fraction unique DHSs')
plt.legend(['all models', '#1', '#2', '#3'])
plt.show()
```
